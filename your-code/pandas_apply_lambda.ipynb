### **Challenge 1. Warm up**
We want to create **bins** of movies according to the number of votes they've received. For that matter, we will create a new column named **'bin'** which will tag every movie as follow:
- From 0 to 999 ==> 1
- From 1000 to 9999 ==> 2
- From 10000 to 99999 ==> 3
- From 100000 to 999999 ==> 4
- More than 1000000 ==> 5 

````python
def movie_votes(votes):
    if votes <= 999:
        return 1
    elif votes >= 1000 and votes <= 9999:
        return 2
    elif votes >= 10000 and votes <= 99999:
        return 3
    elif votes >= 100000 and votes <= 999999:
        return 4
    else:
        return 5
````

````python
movie_df["Bin"] = movie_df.apply(lambda x: movie_votes(x["Votes"]), axis=1)
movie_df.head(10)
````

### **Challenge 2. Using axis concept**
We want to know how much is the revenue per minute for every movie.

````python
def revenue_per_minute(revenue, minute):
    return revenue/minute
````

````python
movie_df["Revenue (Per Minute)"] = movie_df.apply(lambda x: revenue_per_minute(x["Revenue (Millions)"], 
                                                                               x["Runtime (Minutes)"]), axis=1)
movie_df.head(10)
````

### **Challenge 3. Using the lambda**
We want to create a new ranking where we add 1 point if the genre is thriller but subtract 1 point if the genre is comedy.

````python
def new_rating(rating, genre):
    if "Thriller" in genre:
        return rating + 1
    elif "Comedy" in genre:
            return rating - 1      
    else:
        return rating
````

````python
movie_df["New Rating"] = movie_df.apply(lambda x: new_rating(x["Rating"], 
                                                            x["Genre"]), axis=1)
movie_df.head(10)
````

###  **Challenge 4. Now the real stuff**
We want to know if the sum of the ASCII value of every character of the movie title divided by the number of votes retrieve a prime number...remember, prime numbers are integers.

````python
def prime_number(word, votes):
    word_ascii = sum([ord(c) for c in word])
    final_number = int(word_ascii / votes)
    flag = False
    if final_number <= 1:
        return False
    elif final_number > 1:
        for i in range(2, final_number):
            if (final_number % i) == 0:
                flag = True
                break
        if flag:
            return False
        else:
            return True  
````

````python
movie_df["ASCII Title Prime"] = movie_df.apply(lambda x: prime_number(x["Title"], x["Votes"]), axis=1)
movie_df.head(10)
````

###  **Challenge 5. And finally some fantasy**
Feel free to propose your own ranking based in aggregations of at least 3 columns of the dataset.



````python
def recommend(vt, rt, metascore):
    new_rt = rt/10*100
    mean_rt_mtscore = (new_rt + metascore)/2
    if vt <= 10000:
        return("NO")
    elif vt > 10000:
        if mean_rt_mtscore >= 70.0:
            return ("YES")
        else:
            return("NO")
````

````python
movie_df["Recommended"] = movie_df.apply(lambda x: recommend(x["Votes"], x["Rating"], x["Metascore"]), axis=1)
movie_df.head(20)
````
